```python
df_new=df.drop('Car_Name',axis=1)
df_new.head()
```

```python
## enter current year
a=int(input("enter current year:-"))
df_new['current year']=a
df_new
```

```python
df_new['no_year']=df_new['current year'] -  df_new['Year']
df_new
```

```python
## drop Year column
df_new=df_new.drop('Year',axis=1)
```

```python
final_df=pd.get_dummies(df_new,drop_first=True)
final_df.head()
```

```python
final_df=final_df.drop('current year',axis=1)
final_df.head()
```

```python
x=final_df.drop('Selling_Price',axis=1)
y=final_df['Selling_Price']
x.head()
```

```python
y.tail()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
x_train.shape
```

```python
## using all regression algoritms to cheak model score after that we decided the model
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
for i,j in enumerate([RandomForestRegressor,LinearRegression,KNeighborsRegressor,Lasso,DecisionTreeRegressor,SVR]):
  rf=j()
  rf.fit(x_train,y_train)
  y_pred=rf.predict(x_test)
  print(rf.score(x_test,y_test))
```

```python
#after that we selected RandomForestRegressor/DecisionTreeRegressor
rf_r=RandomForestRegressor()
```

```python
n_estimators=[int(x) for x in np.linspace(start=100,stop=1300,num=13)]
n_estimators
```

```python
from sklearn.model_selection import RandomizedSearchCV
#now we have to set parameters of gridsearchcv
# No of trees in random forest
n_estimators = [int(i) for i in np.linspace(start = 100, stop = 1300, num = 13)]
# No of features to consider at every split
max_features = ['auto', 'sqrt']
# Max no of levels in tree
max_depth = [int(i) for i in np.linspace(10, 35, num = 6)]
# Min no of samples required to split a node
min_samples_split = [5,10,15,25,75,100]
# Min no of samples required at each leaf node
min_samples_leaf = [2,4,6,10]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf_grid = RandomizedSearchCV(estimator = rf_r, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_grid.fit(x_train,y_train)
```

```python
rf_grid.best_params_
```

```python
rf_grid.best_score_
```
